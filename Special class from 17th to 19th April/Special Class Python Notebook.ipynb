# Varibale Name Rules

## Python is case-sensitive

```python
var = 10
VAR = "abc"
Var = 0.9
vAr = True

print(var,VAR,Var,vAr)
```

## Cannot start variable names with digits or symbols

```python
12xt = "test"  # any digit is an interger value not string to without quotes to the interpretor
$xt12 = "test"  # symbols have special uses in python hence cannot start the variable name with them
```

```python
# variable names can be started using a letter(uppercase or lowercase) or and underscore

xt12 = "test"
XT12 = "test"
_xt12 = "test"
```

## cannot use opreators symbols in variable names

```python
first_name = "John"  # only underscores can be used

# since the interpretor understands that "-" the minus/hypen symbol is for subtraction cannot use it in variable name
first-name = "John"
first+name = "John"
```

# snake_case
In snake_case the words of your variable name has to be lowercase and seperated by underscore symbol

> this_var_name_is_in_snake_case

# camelCase
In camelCase all letters of the first word is lowercase and rest of the words the first letter only start with uppercase
> thisVarNameIsInCamelCase

> **Important things to remember in variable names:**
- can contain A-Z a-z 0-9 _
- cannot start from digits but can contain digits
- case-sensitive
- no char limit
- cannot use reserved words as variable names
- no space between the words in variable name. either use snake_case or camelCase for multiple words in a variable name

# Fundamental Data Types

- Interger -> Digits without decimal point
- Float -> digits with decimal point
- Char/String -> Letters in the English Alphabet
- None -> Empty/Blank Value
- Boolean -> True(1) | False(0)
- Complex -> Imaginary Numbers -> a + bj

```python
# type() function can tell you what data type is there in the variable
print(type(12))  # Integer
print(type(12.45))  # Float
print(type("this is string"))  # String
print(type(None))  # None
print(type(True))  # Boolean
print(type(2+3j))  # Complex
```

```python
# since python is able to identify the type of data based on the value,
# hence no need to specify a data type for a variable
var = None
print(type(var))
var = 12
print(type(var))
var = 13.09
print(type(var))
var = "Text"
print(type(var))
```

```python
# printing multiple values using comma in print function
var1 = 10
var2 = None
print(type(var1),type(var2))
```
